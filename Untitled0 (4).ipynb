```python
# ==== IMPORT LIBRARIES ====
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
from sklearn.preprocessing import LabelBinarizer

import warnings
warnings.filterwarnings("ignore")
```

```python
# ==== LOAD DATA ====
df = pd.read_csv("Clinical Text Data.csv",encoding='ISO-8859-1')

# ==== SHOW DATA ====
print("Dataset Shape:", df.shape)
print("\nSample Rows:")
print(df.head())

# ==== CHECK FOR NULL ====
print("\nMissing values:\n", df.isnull().sum())

# ==== CLEAN TEXT ====
df['Text'] = df['Text'].astype(str).str.lower()
df['Text'] = df['Text'].str.replace(r"[^a-zA-Z\s]", " ", regex=True)

# ==== TARGET DISTRIBUTION ====
plt.figure(figsize=(10,6))
sns.countplot(y='Label', data=df, order=df['Label'].value_counts().index)
plt.title("Label/Class Distribution")
plt.xlabel("Count")
plt.ylabel("Label")
plt.show()
```

```python
top_labels = df['Label'].value_counts().nlargest(3).index

for label in top_labels:
    text = " ".join(df[df['Label'] == label]['Text'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Word Cloud for Label: {label}")
    plt.axis('off')
    plt.show()
```

```python
# ==== Train/Test Split ====
X = df['Text']
y = df['Label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# ==== Print Shapes ====
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# ==== Show Sample of Training and Test Data ====
print("\n--- Sample Training Data ---")
print(pd.DataFrame({'Text': X_train[:5], 'Label': y_train[:5]}))

print("\n--- Sample Test Data ---")
print(pd.DataFrame({'Text': X_test[:5], 'Label': y_test[:5]}))
```

```python
# ==== STEP 2: FEATURE SELECTION ====
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import chi2, SelectKBest, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Convert text to TF-IDF features
tfidf = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Filter-based: Chi-square test
chi2_selector = SelectKBest(chi2, k=1000)
X_train_chi2 = chi2_selector.fit_transform(X_train_tfidf, y_train)
X_test_chi2 = chi2_selector.transform(X_test_tfidf)

# Wrapper-based: Recursive Feature Elimination (RFE) with Logistic Regression
rfe_estimator = LogisticRegression(max_iter=1000)
rfe_selector = RFE(rfe_estimator, n_features_to_select=500, step=100)
X_train_rfe = rfe_selector.fit_transform(X_train_tfidf, y_train)
X_test_rfe = rfe_selector.transform(X_test_tfidf)

# Embedded: Feature importance from Random Forest
rf_selector = RandomForestClassifier(n_estimators=100, random_state=42)
rf_selector.fit(X_train_chi2, y_train)
importances = rf_selector.feature_importances_

print("Feature selection complete: Chi2, RFE, and RandomForest importance calculated.")
```

```python
model_results = {}  # Store accuracy of each model

def evaluate_model(name, model):
    global model_results

    print(f"\n================ {name} ================\n")

    pipeline = Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english')),
        ('clf', model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    # Accuracy & Report
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")
    model_results[name] = acc  # Save accuracy

    print("Classification Report:\n", classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred, labels=np.unique(y))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y))
    disp.plot(xticks_rotation=90, cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

```python
# Naive Bayes
evaluate_model("Multinomial Naive Bayes", MultinomialNB())

```

```python
# Logistic Regression
evaluate_model("Logistic Regression", LogisticRegression(max_iter=1000))
```

```python

# Support Vector Machine
evaluate_model("Linear SVM", LinearSVC())
```

```python
evaluate_model(
    "Random Forest",
    RandomForestClassifier(
        n_estimators=100,
        max_depth=10,               # controls tree depth
        min_samples_split=10,       # must have at least 10 to split
        min_samples_leaf=5,         # each leaf must have at least 5
        random_state=42
    )
)
```

```python
# === ANN MODEL ===
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

print("\n===== Artificial Neural Network (ANN) =====\n")

# Step 1: Vectorize with TF-IDF (limit dimensions for memory)
tfidf_ann = TfidfVectorizer(stop_words='english', max_features=5000)
X_tfidf = tfidf_ann.fit_transform(X).toarray()

# Step 2: Encode Labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Step 3: Train/Test Split (again, just for ANN)
X_train_ann, X_test_ann, y_train_ann, y_test_ann = train_test_split(
    X_tfidf, y_encoded, stratify=y_encoded, test_size=0.2, random_state=42
)

# Step 4: One-hot encode labels
num_classes = len(np.unique(y_encoded))
y_train_cat = to_categorical(y_train_ann, num_classes)
y_test_cat = to_categorical(y_test_ann, num_classes)

# Step 5: Build ANN Model
model_ann = Sequential()
model_ann.add(Dense(256, input_dim=5000, activation='relu'))
model_ann.add(Dropout(0.3))
model_ann.add(Dense(128, activation='relu'))
model_ann.add(Dropout(0.2))
model_ann.add(Dense(num_classes, activation='softmax'))

model_ann.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 6: Train the ANN Model
history = model_ann.fit(
    X_train_ann, y_train_cat,
    epochs=5, batch_size=64,
    validation_data=(X_test_ann, y_test_cat),
    verbose=1
)

# Step 7: Evaluate ANN
loss, acc = model_ann.evaluate(X_test_ann, y_test_cat, verbose=0)
print(f"\nANN Accuracy: {acc:.4f}")
model_results["ANN"] = acc

# Step 8: Confusion Matrix for ANN
y_pred_ann = model_ann.predict(X_test_ann)
y_pred_classes = np.argmax(y_pred_ann, axis=1)

cm_ann = confusion_matrix(y_test_ann, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_ann, display_labels=le.classes_)
disp.plot(xticks_rotation=90, cmap="Purples")
plt.title("Confusion Matrix - ANN")
plt.show()

# Step 9: Classification Report
print("Classification Report (ANN):\n")
print(classification_report(y_test_ann, y_pred_classes, target_names=le.classes_))
```

```python
# ==== STEP 4: ENSEMBLING & STACKING ====
from sklearn.ensemble import VotingClassifier, StackingClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# Define base models
log_reg = LogisticRegression(max_iter=1000)
svm_model = LinearSVC()
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Voting Classifier
voting_clf = VotingClassifier(
    estimators=[
        ('lr', log_reg),
        ('rf', rf_model),
        ('svm', svm_model)
    ],
    voting='hard'  # Change to 'soft' if using probability-based models
)
voting_clf.fit(X_train_chi2, y_train)

# Stacking Classifier
stacking_clf = StackingClassifier(
    estimators=[
        ('lr', log_reg),
        ('rf', rf_model)
    ],
    final_estimator=LogisticRegression(max_iter=1000)
)
stacking_clf.fit(X_train_chi2, y_train)

# Evaluate ensemble models
voting_pred = voting_clf.predict(X_test_chi2)
stacking_pred = stacking_clf.predict(X_test_chi2)

voting_acc = accuracy_score(y_test, voting_pred)
stacking_acc = accuracy_score(y_test, stacking_pred)

# Store in model_results dictionary (force to float to avoid issues)
model_results["Voting Classifier"] = float(voting_acc)
model_results["Stacking Classifier"] = float(stacking_acc)

# === Accuracy Comparison Plot ===
plt.figure(figsize=(8, 6))

# Filter only valid numeric values for plotting
valid_results = {k: v for k, v in model_results.items() if isinstance(v, (int, float))}
plt.barh(list(valid_results.keys()), list(valid_results.values()), color='skyblue')

plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.title("Model Accuracy Comparison")

# Add text labels for each bar
for i, (k, v) in enumerate(valid_results.items()):
    plt.text(v + 0.002, i, f"{v:.4f}", va='center')

plt.show()

```
